* **Variável preditoras X:** métricas ou não métricas.<br>

**Objetivos da Técnica:** 

* **Atribuição de Probabilidades:** Estimar a probabilidade de ocorrência de determinando evento ou de que uma observação venha a se enquadrar nessa ou
naquela categoria.

* **Classificação em categorias:**
Classificar indivíduos ou observações em categorias específicas.

**Probabilidade do não evento:** A probabilidade do não evento, também conhecida como probabilidade complementar, é a probabilidade de que um evento não ocorra. É calculada subtraindo-se a probabilidade do evento ocorrer de 1.

Por exemplo, suponha que a probabilidade de ocorrer uma chuva em uma determinada cidade seja de 0,3. A probabilidade do não evento (ou seja, de não chover) é de 0,7 (que é 1 - 0,3). Ou seja, a probabilidade de que o evento ocorra (chova) somada à probabilidade do não evento (não chover) resulta em 1.

**Gráfico**

Para plotagem do comportamento da equação, como o logito (z) é uma variável que tende ao infinito, foi proposto um espaço amostral de -5 a 5. Vemos que a função tem uma saída sigmoidal que pode ser chamada também de curva S ou curva logítica.

obs: é importante salientar que o ln é usado para transforma a multiplicação de probabilidades em uma adição, e limitar a observação probabilística de 0 a 1.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definindo a coluna "z" de -5 a 5
z = np.arange(-5, 6)

# Definindo a coluna "p" como logit de "z"
p = 1 / (1 + np.exp(-z))

# Criando o dataframe com as colunas "z" e "p"
df = pd.DataFrame({'z': z, 'p': p})

# Plotando o gráfico de "p" em função de "z"
plt.plot(df['z'], df['p'])
plt.xlabel('z')
plt.ylabel('p')
plt.title('Gráfico de p em função de z')
plt.show()
```

**Conclusão**

A curva logística, ou sigmóide, descreve a relação entre a probabilidade associada à ocorrência de determinando evento e um conjunto de variáveis preditoras. A função logística assume valores entre 0 e 1 para qualquer
Z entre −∞ e +∞.
